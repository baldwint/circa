```python
xgalvo = GalvoPixel("Dev2/ao0", reverse=True)
ygalvo = GalvoPixel("Dev2/ao1")
xgalvo.value = 1880
ygalvo.value = 2100
```

```python
from IPython.html import widgets
from IPython import display

def galvo_ui(n=5, origin=None):
    
    if origin is None:
        origin = (xgalvo.value, ygalvo.value)

    def make_move(galvo, n):
        def move(inst):
            galvo.set_value(galvo.get_value() + n)
        return move

    def gen_widgets():
        for galvolabel,galvo in zip('xy', (xgalvo, ygalvo)):
            for dirlabel,direction in zip('+-', (+1,-1)):
                label = galvolabel + dirlabel
                w = widgets.ButtonWidget(description=label)
                w.on_click(make_move(galvo, direction * n))
                yield w

    for w in gen_widgets():
        display.display(w)
    
    w = widgets.ButtonWidget(description='0')
    w.on_click(lambda inst: (ygalvo.set_value(origin[1]), xgalvo.set_value(origin[0])))
    display.display(w)
```

##Imaging

```python
def chunk(gen, n=1):
    # not perfect unless n perfectly divides the lenght of gen
    while True:
        for i in xrange(n-1):
            next(gen)
        yield next(gen)
```

```python
import itertools

def yielding_fromiter(gen, result):
    """
    fills the ND array ``result`` with values from the ``gen`` iterator,
    yielding after each insertion so that you can do something else
    (like incremental plotting).
    
    """
    dimensions = [xrange(n) for n in result.shape]
    coords = itertools.product(*dimensions)
    for loc,val in itertools.izip(coords,gen):
        result[loc] = val
        yield
```

```python
def gen2D(X, Y):
    for y in Y:
        ygalvo.value = y
        for x in X:
            xgalvo.value = x
            yield
            
def scan(gen, t=0.1):
    """threaded version"""
    p,c = configure_counter(duration=t)
    with counting(p,c):
        next(gen)
        last = do_count(p,c)/t
        for step in gen:
            start_count(p, c)
            yield last
            last = finish_count(p, c)/t
        yield last
```

```python
def make_array(fn, **kwargs):
    """
    makes a memory-mapped file array, complete with .npy header
    """
    # create result array on disk
    a = n.ndarray(**kwargs)
    default = n.nan
    a[:] = default
    n.save(fn, a)
    del a

    # reopen as memory mapped array
    vector = n.load(fn, mmap_mode='r+')
    return vector
```

```python
from IPython import display
import matplotlib as mpl

def plotgen2D(gen, vector, ax=None, im=None):
    """
    Plot (imshow) the 2D array ``vector``,
    updating the plot every time ``gen`` yields.
    It is assumed that ``gen`` is modifying ``vector``.
    
    Will do an imshow with scalebars to the given ``ax``,
    unless a preexisting ``im`` is provided.
    """
    
    if ax is None:
        ax = gca()
    if im is None:
        im = ax.imshow(vector)
        cbar = ax.figure.colorbar(im)
    for img in gen:
        im.set_data(vector)
        im.autoscale()
        if 'inline' in mpl.get_backend():
            display.clear_output(wait=True)
            display.display(ax.figure)
        else:
            ax.figure.canvas.draw()
```

```python
from scalebars import add_scalebar

def show_image(vector, X, Y,
               cal = None, bar = 1.,
               ax = None, **kwargs):
    """
    show a density plot with a scalebar,
    
    cal: calibration (um per pixel)
    bar: length of bar (in um)
    
    """
    if ax is None:
        ax = gca()
    im = density_plot(vector, X, Y, ax=ax, **kwargs)
    cbar = ax.figure.colorbar(im)
    if cal is not None:
        sb = add_scalebar(ax,
                          sizex=bar/cal,
                          labelx='%dum' % bar,
                          matchx=False, matchy=False,
                          hidex=False, hidey=False,
                          pad=.3, borderpad=.2,
                          frameon=True)
        sb.patch.set_alpha(0.5)
    return im
```

```python
import itertools
c = itertools.count(1)
```

```python
def do_scan(X, Y, t=0.01, bar=10.):
    i = next(c)
    fn = 'datafile%i.npy' % i
    vector = make_array(fn, shape=Y.shape + X.shape, dtype=float)

    im = show_image(vector, X, Y, cal=.083, bar=bar) # 100x objective
    # for 40x objective: .172

    # do the scan
    gen = chunk(yielding_fromiter(scan(gen2D(X, Y), t=t), vector),
                n = vector.shape[-1])
    with notraceback():
        plotgen2D(gen, vector, im=im)

    z_fn = 'image%d.npz' % i
    n.savez(z_fn, image=vector, X=X, Y=Y)

    print fn, z_fn
```
